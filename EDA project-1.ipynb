```python
#Finding the outliers in each numerical column
import pandas as pd
df = pd.read_excel('C:\\Users\\ladis\\Downloads\\data.xlsx', engine='openpyxl')
# Function to find outliers using IQR
def find_outliers_IQR(data):
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    outliers = ((data < (Q1 - 1.5 * IQR)) | (data > (Q3 + 1.5 * IQR)))
    return outliers

# Apply the function to each numerical column
numerical_columns = df.select_dtypes(include=['float64', 'int64']).columns
outliers = df[numerical_columns].apply(find_outliers_IQR)
print(outliers)
```

```python
# probability and frequency distribution of each numerical column
import matplotlib.pyplot as plt
import seaborn as sns
for column in numerical_columns:
    plt.figure(figsize=(10, 6))
    sns.histplot(df[column], kde=True)
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```

```python
relevant_jobs = [
    'programmer analyst',         # For Programming Analyst
    'program analyst trainee',     # Alternate for Programming Analyst
    'software engineer',           # For Software Engineer
    'senior software engineer',    # Alternate for Software Engineer
    'software engineer trainee',    # Alternate for Software Engineer
    'hardware engineer',           # For Hardware Engineer
    'associate engineer',          # For Associate Engineer
    'associate software developer'  # Alternate for Associate Engineer
]
import pandas as pd
from scipy import stats

# Load the data
df = pd.read_excel('C:\\Users\\ladis\\Downloads\\data.xlsx', engine='openpyxl')

# Explore the data
print(df.head())

# Filter for relevant job titles
relevant_jobs = [
    'programmer analyst',
    'program analyst trainee',
    'software engineer',
    'senior software engineer',
    'software engineer trainee',
    'hardware engineer',
    'associate engineer',
    'associate software developer'
]
filtered_data = df[df['Designation'].isin(relevant_jobs)]

# Check if filtered_data is empty
if filtered_data.empty:
    print("No data available for the specified job titles.")
else:
    # Calculate mean and median salaries
    mean_salary = filtered_data['Salary'].mean() / 100000  # Convert to Lakhs
    median_salary = filtered_data['Salary'].median() / 100000  # Convert to Lakhs

    # Check how many earn in the specified range
    within_range = filtered_data[(filtered_data['Salary'] >= 250000) & (filtered_data['Salary'] <= 300000)]

    # Statistical Test (One-sample t-test)
    t_statistic, p_value = stats.ttest_1samp(filtered_data['Salary'], 275000)  # Use the midpoint of the range

    # Print results
    print(f"Mean Salary: {mean_salary:.2f} Lakhs")
    print(f"Median Salary: {median_salary:.2f} Lakhs")
    print(f"Number of graduates within the range: {len(within_range)}")
    print(f"T-statistic: {t_statistic:.2f}, P-value: {p_value:.4f}")

    # Conclusion based on P-value
    if p_value < 0.05:
        print("Reject the null hypothesis: The average salary is significantly different from the claimed range.")
    else:
        print("Fail to reject the null hypothesis: The average salary is not significantly different from the claimed range.")
```

```python
import pandas as pd
import scipy.stats as stats

# Create a contingency table
contingency_table = pd.crosstab(df['Gender'], df['Specialization'])

# Perform the chi-square test
chi2, p, dof, expected = stats.chi2_contingency(contingency_table)

# Print the results
print(f"Chi-square statistic: {chi2:.2f}")
print(f"P-value: {p:.4f}")

# Interpretation
if p < 0.05:
    print("There is a significant relationship between gender and specialization.")
else:
    print("There is no significant relationship between gender and specialization.")
```
